```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
df.sample(4)
```

```python
df['Survived'].value_counts()
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x = df['Survived'].value_counts().index, y = df['Survived'].value_counts())
plt.title('Value Counts of Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Contar valores por clase y supervivencia
datos = df.groupby(['Pclass', 'Survived']).size().reset_index(name='Total')

# Crear grfica
sns.barplot(data=datos, x='Pclass', y='Total', hue='Survived', palette=['#2d1e3e', '#5a7d8c'])

plt.title('Tasa de Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Contar valores por sexo y supervivencia
datos = df.groupby(['Sex', 'Survived']).size().reset_index(name='Total')

# Crear grfica
sns.barplot(data=datos, x='Sex', y='Total', hue='Survived', palette=['#2d1e3e', '#5a7d8c'])

plt.title('Tasa de Supervivencia por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.tight_layout()
plt.show()
```

```python
# Crear categoras de edad
def categorizar_edad(edad):
    if edad <= 2:
        return 'Infante'
    elif edad <= 12:
        return 'Nio'
    elif edad <= 17:
        return 'Adolescente'
    elif edad <= 35:
        return 'Joven Adulto'
    elif edad <= 60:
        return 'Adulto'
    else:
        return 'Mayor'

# Aplicar la funcin a la columna Age
df['Grupo_Edad'] = df['Age'].apply(categorizar_edad)

# Crear la grfica
plt.figure(figsize=(12, 6))

# Contar valores por grupo de edad y supervivencia
datos = df.groupby(['Grupo_Edad', 'Survived']).size().reset_index(name='Total')

# Definir el orden de las categoras
orden = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']

sns.barplot(data=datos, x='Grupo_Edad', y='Total', hue='Survived', 
            palette=['#2d1e3e', '#5a7d8c'], order=orden)

plt.title('Tasa de Supervivencia por Grupo de Edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Total')
plt.legend(title='Supervivencia', labels=['No sobrevivi', 'Sobrevivi'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calcular porcentaje de supervivencia por SibSp
supervivencia_sibsp = df.groupby('SibSp')['Survived'].agg(['sum', 'count'])
supervivencia_sibsp['porcentaje'] = (supervivencia_sibsp['sum'] / supervivencia_sibsp['count'] * 100).round(2)
supervivencia_sibsp.columns = ['Sobrevivientes', 'Total', 'Porcentaje (%)']

print("Porcentaje de Supervivencia por Nmero de Hermanos/Esposos a Bordo (SibSp):\n")
print(supervivencia_sibsp)
```

```python
plt.figure(figsize=(12, 8))

# Seleccionar solo columnas numricas
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns

# Calcular matriz de correlacin
correlacion = df[columnas_numericas].corr()

# Crear mapa de calor
sns.heatmap(correlacion, 
            annot=True, 
            fmt='.2f',
            cmap='coolwarm', 
            center=0,
            square=True,
            linewidths=1,
            cbar_kws={'label': 'Nivel de Correlacin'})

plt.title('Matriz de Correlacin de Variables Numricas')
plt.tight_layout()
plt.show()
```
